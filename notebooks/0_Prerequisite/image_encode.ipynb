```python
import matplotlib.pyplot as plt
import tensorflow as tf
image_raw_data = tf.gfile.FastGFile("/home/dev/gitcode/transfer_learning/flower_photos/daisy/4654579740_6671a53627_m.jpg","r").read()
#print image_raw_data
with tf.Session() as sess:
    image_data = tf.image.decode_jpeg(image_raw_data)
    print image_data.eval()
    plt.imshow(image_data.eval())
    plt.show()
    image_data = tf.image.convert_image_dtype(image_data,dtype=tf.uint8)
    encoded_image = tf.image.encode_jpeg(image_data)
    with tf.gfile.GFile('/tmp/4654579740_6671a53627_m.jpg','wb') as f:
        f.write(encoded_image.eval())
```
